- ### Subtask 2.3: The General Audience and the Critics

You might have noticed the column `MetaCritic` in this dataset. This is a very popular website where an average score is determined through the scores given by the top-rated critics. Second, you also have another column `IMDb_rating` which tells you the IMDb rating of a movie. This rating is determined by taking the average of hundred-thousands of ratings from the general audience. 

As a part of this subtask, you are required to find out the highest rated movies which have been liked by critics and audiences alike.
1. Firstly you will notice that the `MetaCritic` score is on a scale of `100` whereas the `IMDb_rating` is on a scale of 10. First convert the `MetaCritic` column to a scale of 10.
2. Now, to find out the movies which have been liked by both critics and audiences alike and also have a high rating overall, you need to -
    - Create a new column `Avg_rating` which will have the average of the `MetaCritic` and `Rating` columns
    - Retain only the movies in which the absolute difference(using abs() function) between the `IMDb_rating` and `Metacritic` columns is less than 0.5. Refer to this link to know how abs() funtion works - https://www.geeksforgeeks.org/abs-in-python/ .
    - Sort these values in a descending order of `Avg_rating` and retain only the movies with a rating equal to higher than `8` and store these movies in a new dataframe `UniversalAcclaim`.
    

```python
# Changing the scale of 'MetaCritic' from 100 to 10

movies["MetaCritic"] = movies["MetaCritic"] / 10
```

```python
# Finding the average ratings of 'MetaCritic' & 'IMDb_rating' columns and storing them in a new column 'Avg_rating'

movies["Avg_rating"] = ( movies["MetaCritic"] + movies["IMDb_rating"] ) / 2
```

```python
# Finding the movies with absolute difference between 'MetaCritic' and 'Avg_rating' < 0.5 
# & also with the average rating of >= 8 and storing these movies in 'UniversalAcclaim'

UniversalAcclaim = movies[(abs(movies["MetaCritic"] - movies["IMDb_rating"]) < 0.5) & (movies["Avg_rating"] >= 8)]
```

```python
# Sorting these values in a descending order of 'Avg_rating'

UniversalAcclaim.sort_values(by = "Avg_rating", ascending = False, inplace = True, ignore_index = True)

UniversalAcclaim
```

-  ###  Subtask 3.1 Combine the Dataframe by Genres

There are 3 columns in the dataframe - `genre_1`, `genre_2`, and `genre_3`. As a part of this subtask, you need to aggregate a few values over these 3 columns. 
1. First create a new dataframe `df_by_genre` that contains `genre_1`, `genre_2`, and `genre_3` and all the columns related to **CVotes/Votes** from the `movies` data frame. There are 47 columns to be extracted in total.
2. Now, Add a column called `cnt` to the dataframe `df_by_genre` and initialize it to one. You will realise the use of this column by the end of this subtask.
3. First group the dataframe `df_by_genre` by `genre_1` and find the sum of all the numeric columns such as `cnt`, columns related to CVotes and Votes columns and store it in a dataframe `df_by_g1`.
4. Perform the same operation for `genre_2` and `genre_3` and store it dataframes `df_by_g2` and `df_by_g3` respectively. 
5. Now that you have 3 dataframes performed by grouping over `genre_1`, `genre_2`, and `genre_3` separately, it's time to combine them. For this, add the three dataframes and store it in a new dataframe `df_add`, so that the corresponding values of Votes/CVotes get added for each genre.There is a function called `add()` in pandas which lets you do this. You can refer to this link to see how this function works. https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.add.html
6. The column `cnt` on aggregation has basically kept the track of the number of occurences of each genre.Subset the genres that have atleast 10 movies into a new dataframe `genre_top10` based on the `cnt` column value.
7. Now, take the mean of all the numeric columns by dividing them with the column value `cnt` and store it back to the same dataframe. We will be using this dataframe for further analysis in this task unless it is explicitly mentioned to use the dataframe `movies`.
8. Since the number of votes can't be a fraction, type cast all the CVotes related columns to integers. Also, round off all the Votes related columns upto two digits after the decimal point.


```python
# Viewing the list of columns in 'movies' dataset just to understand the order in which columns are stored

movies.columns
```

```python
# Creating the dataframe 'df_by_genre' that has all 'genre' columns, 'CVotes' columns and 'Votes' columns

df_by_genre = movies.loc[:,'genre_1':'genre_3'].join(movies.loc[:,'CVotes10':'VotesnUS'])

df_by_genre.head()
```

```python
# Creating a column cnt and initializing it to 1

df_by_genre["cnt"] = 1
```

```python
# Grouping and summing the movies by individual genres and creating a dataframe for each

df_by_g1 = df_by_genre.groupby(by = ["genre_1"]).sum()
df_by_g2 = df_by_genre.groupby(by = ["genre_2"]).sum()
df_by_g3 = df_by_genre.groupby(by = ["genre_3"]).sum()
```

```python
# Adding all the grouped data frames and storing them in a new data frame 'df_add'

df_add = df_by_g1.add(df_by_g2, fill_value = 0).add(df_by_g3, fill_value = 0)

df_add
```

```python
# Extracting genres with atleast 10 occurences with the help of the 'cnt' column, which, on aggregation 
# has basically kept the track of the number of occurences of each genre.
# Updating the 'genre_top10' with these values

genre_top10 = df_add[df_add["cnt"] >= 10]

genre_top10
```

```python
# Taking the mean for every column by dividing with cnt 

genre_top10.loc[:,'CVotes10':'VotesnUS'] = genre_top10.loc[:,'CVotes10':'VotesnUS'].apply(lambda x: x/genre_top10["cnt"], axis = 0)

genre_top10.head()
```

```python
# Rounding off the columns of 'Votes' to two decimals

genre_top10.loc[:,'VotesM':'VotesnUS'] = genre_top10.loc[:,'VotesM':'VotesnUS'].round(2)
```

```python
genre_top10.head()
```

```python
# Converting 'CVotes' columns to int type

CVotes_columns = ['CVotes10', 'CVotes09', 'CVotes08', 'CVotes07', 'CVotes06', 'CVotes05', 'CVotes04', 'CVotes03', 
             'CVotes02', 'CVotes01', 'CVotesMale', 'CVotesFemale', 'CVotesU18', 'CVotesU18M', 'CVotesU18F', 
             'CVotes1829', 'CVotes1829M', 'CVotes1829F', 'CVotes3044', 'CVotes3044M', 'CVotes3044F', 'CVotes45A', 
             'CVotes45AM', 'CVotes45AF', 'CVotes1000', 'CVotesUS', 
             'CVotesnUS']

genre_top10[CVotes_columns] = genre_top10[CVotes_columns].astype(int)

genre_top10.info()
```

The dataset contains both the US and non-US movies. Let's analyse how both the US and the non-US voters have responded to the US and the non-US movies.

1. Create a column `IFUS` in the dataframe `movies`. The column `IFUS` should contain the value "USA" if the `Country` of the movie is "USA". For all other countries other than the USA, `IFUS` should contain the value `non-USA`.


2. Now make a boxplot that shows how the number of votes from the US people i.e. `CVotesUS` is varying for the US and non-US movies. Make use of the column `IFUS` to make this plot. Similarly, make another subplot that shows how non US voters have voted for the US and non-US movies by plotting `CVotesnUS` for both the US and non-US movies. Write any of your two inferences/observations from these plots.


3. Again do a similar analysis but with the ratings. Make a boxplot that shows how the ratings from the US people i.e. `VotesUS` is varying for the US and non-US movies. Similarly, make another subplot that shows how `VotesnUS` is varying for the US and non-US movies. Write any of your two inferences/observations from these plots.

Note : Use `movies` dataframe for this subtask. Make use of this documention to format your boxplot - https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
# Creating 'IFUS' which contains the value 'USA' if the Country of the movie is USA. 
# For all other countries other than the USA, it contains the value 'non-USA'.

movies["IFUS"] = movies["Country"].apply(lambda x: "USA" if x=="USA" else "non-USA")

movies.head()
```

```python
# Box plot 1 shows how the number of votes from the USA people are varying for the USA and non-USA movies
# Box plot 2 shows how the number of votes from the non-USA people are varying for the USA and non-USA movies

plt.figure(figsize = [20,6])


plt.subplot(1,2,1)
sns.boxplot(data = movies, y = "CVotesUS", x = "IFUS", palette="colorblind")

plt.title("Box plot of number of Votes from people of USA \nfor movies made in USA & other countries\n", 
          fontdict = {"fontsize":15, "color":"tab:blue"})
plt.xlabel("Country of origin of movie", fontdict = {"fontsize":12, "color":"tab:blue"})
plt.ylabel("Number of votes from people of USA\n", fontdict = {"fontsize":12, "color":"tab:blue"})

plt.subplot(1,2,2)
sns.boxplot(data = movies, y = "CVotesnUS", x = "IFUS", palette="colorblind")

plt.title("Box plot of number of Votes from people of other countries\n for movies made in USA & other countries\n",
         fontdict = {"fontsize":15, "color":"goldenrod"})
plt.xlabel("Country of origin of movie", fontdict = {"fontsize":13, "color":"goldenrod"})
plt.ylabel("Number of votes from people of other countries\n", fontdict = {"fontsize":13, "color":"goldenrod"})


plt.show()
```

**`Inferences:`** Write your two inferences/observations below:

**- Inference 1:** It appears that people of USA as well as people of other countries have voted more for USA-originated movies (since the upper fences of both the USA-originated movies' boxplots are much higher than those of the non-USA-originated movies' plots), thus we can hypothesise that Hollywood movies are popular in all countries compared to non-USA originated movies.

**- Inference 2:** Since the medians (50th percentiles) of both the boxplots in the second subplot appear to be very similar, we can infer that the first 50% votes for both kinds of origins, non-USA-originated movies and USA-originated movies, are similar from the people of other countries.

```python
# Box plot 3 shows how the average ratings from the USA people are varying for the USA and non-USA movies
# Box plot 4 shows how the average ratings from the non-USA people are varying for the USA and non-USA movies

plt.figure(figsize = [20,6])


plt.subplot(1,2,1)
sns.boxplot(data = movies, y = "VotesUS", x = "IFUS", palette="colorblind")

plt.title("Average ratings by people of USA \nfor movies made in USA vs other countries\n",
         fontdict = {"fontsize":15, "color":"tab:blue"})
plt.xlabel("Country of origin of movie", fontdict = {"fontsize":12, "color":"tab:blue"})
plt.ylabel("Average rating by people of USA\n", fontdict = {"fontsize":12, "color":"tab:blue"})


plt.subplot(1,2,2)
sns.boxplot(data = movies, y = "VotesnUS", x = "IFUS", palette="colorblind")

plt.title("Average ratings by people of other countries \nfor movies made in USA vs other countries\n",
         fontdict = {"fontsize":15, "color":"goldenrod"})
plt.xlabel("Country of origin of movie", fontdict = {"fontsize":12, "color":"goldenrod"})
plt.ylabel("Average rating by people of other countries\n", fontdict = {"fontsize":12, "color":"goldenrod"})

plt.show()
```

**`Inferences:`** Write your two inferences/observations below:

**- Inference 1:** It appears that both sets of movie raters (prople of USA and other countries) have given highier average ratings to movies that have originated in the USA since the upper fences of both boxplots of USA-originated movies are highier than those of movies originated in other countries. We can hypothesize that Hollywood movies are well liked by people of all countries.

**- Inference 2:** It appears, in both cases, that there is a greater variance in the kinds of ratings given to USA-originated movies, as compared to non-USA-originated-movies (which seem to have more concentrated ratings). In the first subplot (people of USA), the range of average ratings for USA-originated movies extends from 7.5 to 8.5 but for non-USA-originated-movies its only from 7.5 to 8.2. In the second subplot (people of other countries) we see the same pattern, the range of average ratings for USA-originated movies extends from 7.3 to 8.3 but for non-USA-originated-movies its only from 7.3 to 8.1.
